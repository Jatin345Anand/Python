```python
def RemoveSID(s1):
#     print('sid = ',s1.find('SID:') )
    return s1[s1.index(':',0,len(s1))+2:len(s1)]

def RemoveSchemaID(s2):
    return s2[s2.index(':',0,len(s2))+2:len(s2)]    
def RemoveBrakets(S):
    A=[]
    i1 = S.find("(")
    i2 = S.find(")")
    A.append(S[0:i1])
    A.append(S[i1+1:i2])
    return A
    
def RemoveSpaces(S):
    if(S.find(" ")<0):
        return S
    else:
        A=[]
        S1=S.split(" ")
        for i in range(len(S1)):
            if(len(S1[i])>0):
                A.append(S1[i])
        return A

 

def CreateEfficientDatafromRawofStudent(L):
    EfL=[]
    EfL.append(L[0])
    EfL.append(L[2])
    if(L[4].find("SID:")>-1):
        EfL.append(RemoveSID(L[4]))
    else:
        EfL.append(L[4])
    if(L[6].find("SchemeID:")>-1):
        EfL.append(RemoveSchemaID(L[6]))
    else:
        EfL.append(L[6])
    for i in range(7,len(L)):
        if(L[i].find('AA')>-1 or L[i].find('DD')>-1 or L[i].find('CC')>-1):
            for i in range(4):
                EfL.append('A')
            continue
        if(L[i].find('(')>-1):
            if('A' in L[i] or 'D' in L[i] or 'C' in L[i] or 'L' in L[i]):
                L[i]=L[i][1:]
            BL= RemoveBrakets(L[i])
            for k2 in range(len(BL)):
                EfL.append(BL[k2])
            continue
        if(L[i].find(' ')>-1):
            SL= RemoveSpaces(L[i])
            for k1 in range(len(SL)):
                EfL.append(SL[k1])
            continue
    EfL.append('00') 
    return EfL
```
