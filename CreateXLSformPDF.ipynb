```python
def CreatLforOnePlayer(L):
    LEN = len(L)
    ans=[]
    print(L)
    indexArrayofBrackets=[]
    for j in range(1,len(L)):
        if(L[j].find('(')>-1 or L[j].find(')')>-1):
            if(L[j][1:-1].isdigit()):
                indexArrayofBrackets.append(j)
    print(indexArrayofBrackets)
    if(len(indexArrayofBrackets)==1):
        if(L[0].find('(-)')>-1):
            ans.append(0)
        else:
            ans.append(int(L[0][1:]))
        ans.append(int(L[1]))
        nameP=''
        for k in range(2,indexArrayofBrackets[0]-1):
            nameP =nameP + L[k] + ' '
        nameP = nameP + L[indexArrayofBrackets[0]-1]
        ans.append(nameP)
        ans.append(int(L[indexArrayofBrackets[0]][1:-1]))
        print('HAlf L = ',ans)
        countyP=''
        if(len(L[-1])==1):
            for x in range(indexArrayofBrackets[0]+1,len(L)-3):
                countyP = countyP + L[x] + ' '
            countyP = countyP + L[len(L)-3]
            ans.append(countyP)
            ans.append(int(L[-2]))
            ans.append(int(L[-1]))
        if(len(L[-1])>1):
            for x in range(indexArrayofBrackets[0]+1,len(L)-2):
                countyP = countyP + L[x] + ' '
            countyP = countyP + L[len(L)-2]
            ans.append(countyP)
            ans.append(int(L[-1]))
            ans.append(0)
    if(len(indexArrayofBrackets)==2):
        ans1=[]
        ans2=[]
        midRowIndex=0
        for i1 in range(1,len(L)):
            if(L[i1].find('-')>-1):
                midRowIndex= i1
                break
        print(midRowIndex,L[midRowIndex],'%%',indexArrayofBrackets)
        if(L[0].find('(-)')>-1):
            ans1.append(0)
        else:
            ans1.append(int(L[0][1:]))
        ans1.append(int(L[1]))
        nameP1=''
        nameP2=''
        countyP1=''
        countyP2=''
        for j1 in range(2,indexArrayofBrackets[0]-1):
            nameP1 = nameP1 + L[j1] + ' '
        nameP1 = nameP1 + L[indexArrayofBrackets[0]-1]
        ans1.append(nameP1)
        ans1.append(int(L[indexArrayofBrackets[0]][1:-1]))
        LastStringofP1 = L[midRowIndex][:L[midRowIndex].find('-')]
        FirstStringofP2 = L[midRowIndex][L[midRowIndex].find('-')+1:]
        print(ans1,LastStringofP1,FirstStringofP2)
        Difference = midRowIndex - indexArrayofBrackets[0]
        print(Difference,midRowIndex,indexArrayofBrackets[0])
        if(Difference==2):
            print('In difference 1....')
            ans1.append(L[midRowIndex-1])
            print(L[midRowIndex-1])
            ans1.append(int(LastStringofP1))
            ans1.append(0)
        if(Difference==3):
            print('In Differenc 2....')
            if(len(LastStringofP1)==1):
                ans1.append(L[midRowIndex-2])
                ans1.append(int(L[midRowIndex-1]))
                ans1.append(int(LastStringofP1))
            if(len(LastStringofP1)>1):
                ans1.append(L[midRowIndex-2]+ ' '+ L[midRowIndex-1])
                ans1.append(int(LastStringofP1))
                ans1.append(0)
        if(Difference>3):
            print('In difference morethan 2...')
            if(len(LastStringofP1)==1):
                for d1 in range(indexArrayofBrackets[0]+1,midRowIndex-2):
                    countyP1 = countyP1+L[d1]+' '
                countyP1 = countyP1 + L[midRowIndex-2]
                ans1.append(countyP1)
                ans1.append(int(L[midRowIndex-1]))
                ans1.append(int(LastStringofP1))
            if(len(LastStringofP1)>1):
                for d1 in range(indexArrayofBrackets[0]+1,midRowIndex-1):
                    countyP1 = countyP1 + L[d1]+ ' '
                countyP1 = countyP1+L[midRowIndex-1]
                ans1.append(countyP1)
                ans1.append(int(LastStringofP1))
                ans1.append(0)
        print('Ans1 = ',ans1)
        print('First String of P2 ',FirstStringofP2,'Last of P1 ',LastStringofP1,'Mid = ',L[11])
        if(FirstStringofP2.find('(-)')>-1):
            ans2.append(0)
        else:
            ans2.append(int(FirstStringofP2))
        ans2.append(int(L[midRowIndex+1]))
        for j1 in range(midRowIndex+2,indexArrayofBrackets[1]-1):
            nameP2 = nameP2 + L[j1] + ' '
        nameP2 = nameP2 + L[indexArrayofBrackets[1]-1]
        ans2.append(nameP2)
        ans2.append(int(L[indexArrayofBrackets[1]][1:-1]))

        Difference2 = (len(L)-1) - indexArrayofBrackets[1]
        print(Difference2,len(L),indexArrayofBrackets[1])
        if(Difference2==2):
            print('In difference 1....')
            ans2.append(L[-2])
            print(L[-2])
            ans2.append(int(L[-1]))
            ans2.append(0)
        if(Difference2==3):
            print('In Differenc 2....')
            if(len(L[-1])==1):
                ans2.append(L[-3])
                ans2.append(int(L[-2]))
                ans2.append(int(L[-1]))
            if(len(L[-1])>1):
                ans2.append(L[-3]+ ' '+ L[-2])
                ans2.append(int(L[-1]))
                ans2.append(0)
        if(Difference2>3):
            print('In difference morethan 2...')
            if(len(L[-1])==1):
                for d1 in range(indexArrayofBrackets[1]+1,len(L)-3):
                    countyP2 = countyP2+L[d1]+' '
                countyP2 = countyP2 + L[-3]
                ans2.append(countyP1)
                ans2.append(int(L[-2]))
                ans2.append(int(L[-1]))
            if(len(L[-1])>1):
                for d1 in range(indexArrayofBrackets[1]+1,len(L)-2):
                    countyP2 = countyP2 + L[d1]+ ' '
                countyP2 = countyP2+L[midRowIndex-1]
                ans2.append(countyP1)
                ans2.append(int(L[-1]))
                ans2.append(0)


        print('Ans2 = ',ans2)
        ans.append(ans1)
        ans.append(ans2)
    return ans
        
```
